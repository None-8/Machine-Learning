```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('student-por.csv')
df
```

```python
print(df.columns)
```

```python
s=np.random.rand(100)*100
plt.scatter(x=df.studytime, y=df.G1, s=s, c='blue',alpha=0.05)
plt.show()
```

```python
plt.scatter(x=df.studytime, y=df.G2, s=s, c='red',alpha=0.05)
plt.show()
```

```python
plt.scatter(x=df.studytime, y=df.G3, s=s, c='green',alpha=0.05)
plt.show()
```
